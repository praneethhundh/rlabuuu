```python
ex-1 :

x<-as.integer(strsplit(readline(prompt="Enter numbers")," ")[[1]]) 
min(x) 
mean(x) 
median(x) 
sqrt(x) 
round(x) 
sum(x) 
avg<-sum(x)/length(x) 
print(avg) 


ex-2:

Import CSV 
data1<-read.table(file.choose(),header=T,sep=".") 
Import Text 
data1<-read.delim(file.choose(),header=T,sep=".") 
EXPORT Text 

df = data.frame( 
"Name" = c("Amiya", "Raj", "Asish"), 
"Language" = c("R", "Python", "Java"), 
"Age" = c(22, 25, 45) 
) 

write.table(df, 
file = "myDataFrame.txt", 
sep = "\t", 
row.names = TRUE, 
col.names = NA) 
EXPORT CSV 

df = data.frame( 
"Name" = c("Amiya", "Raj", "Asish"), 
"Language" = c("R", "Python", "Java"), 
"Age" = c(22, 25, 45) 
) 

write.csv(df, file = "my_data.csv")


ex-3:
 
m1 = matrix(c(1, 2, 3, 4, 5, 6), nrow = 2) 
print("Matrix-1:") 
print(m1) 
m2 = matrix(c(0, 1, 2, 3, 0, 2), nrow = 2) 
print("Matrix-2:") 
print(m2) 
result = m1 + m2 
print("Result of addition") 
print(result) 
result = m1 - m2 
print("Result of subtraction") 
print(result) 
result = m1 %*% m2 
print("Result of multiplication") 
print(result) 
result = m1 / m2 
print("Result of division:") 
print(result)


A <- matrix(c(2, 3, 1,
              4, 7, 5,
              6, 9, 8), 
            nrow = 3, byrow = TRUE)

print("Original Matrix A:")
print(A)

A_transpose <- t(A)
print("Transpose of A:")
print(A_transpose)

A_transpose_inverse <- solve(A_transpose)
print("Inverse of Transpose of A:")
print(A_transpose_inverse)


ex-4:

Mean 
x <- c(1,2,3,4,5,1,2,3,1,2,4,5,2,3,1,1,2,3,5,6) 
mean.result = mean(x) 
print (mean.result) 
Median 
x <- c(1,2,3,4,5,1,2,3,1,2,4,5,2,3,1,1,2,3,5,6) 
median.result = median(x) 
print (median.result) 
Mode: 
mode <- function(x) { 
ux <- unique(x) 
ux[which.max(tabulate(match(x, ux)))] 
} 
x <- c(1,2,3,4,5,1,2,3,1,2,4,5,2,3,1,1,2,3,5,6) 
mode.result = mode(x)  
print (mode.result) 
Standard Deviation 
sd.result = sqrt(var(x)) 
print (sd.result)

ex 5:

a.
install.packages("dplyr") 
library(dplyr) 
install.packages("tidyr") 
library(tidyr) 

data <- read.csv("my_data.csv") 
print(data) 
 
print(sum(is.na(data))) 
 
data <- data %>% drop_na() 

print(data) 


b.
data <- data.frame( 
  x = c(1, 2, 3, 4, 5), 
  y = c(10, 20, 30, 40, 50) 
) 

write.csv(data, "dataset.csv", row.names = FALSE) 

data <- read.csv("dataset.csv") 

min_max_normalize <- function(x) { 
  return ((x - min(x)) / (max(x) - min(x))) 
} 

normalized_data <- as.data.frame(lapply(data, min_max_normalize)) 

print(normalized_data)


ex-6:

library(tidyverse)


house_data <- read.csv("https://raw.githubusercontent.com/Giridharan79/HOUSE_PRICE_PREDICTION/refs/heads/main/final_dataset.csv", header = TRUE)

numeric_data <- house_data %>% select(where(is.numeric))

numeric_data <- numeric_data %>% drop_na()

scaled_data <- scale(numeric_data)

pca_result <- prcomp(scaled_data, center = TRUE, scale. = TRUE)

summary(pca_result)

print(pca_result$rotation)

head(pca_result$x)

plot(pca_result, type = "l", main = "Scree Plot - PCA")

biplot(pca_result, scale = 0)


ex-7:

data(mtcars) 

str(mtcars) 

linear_model <- lm(mpg ~ hp, data = mtcars) 

summary(linear_model) 

plot(mtcars$hp, mtcars$mpg, main = "Simple Linear Regression", xlab = "Horsepower", 
ylab = "Miles Per Gallon") 
abline(linear_model, col = "red") 

new_data <- data.frame(hp = c(100, 200)) 
predictions <- predict(linear_model, newdata = new_data) 
print("Predictions:") 
print(predictions) 

ex-8:


data(iris) 

iris_data <- iris[, -5] 

k <- 3 

kmeans_model <- kmeans(iris_data, centers = k, nstart = 20) 

library(cluster) 
clusplot(iris_data, kmeans_model$cluster, color = TRUE, shade = TRUE, labels = 2, lines = 
0) 

print("Cluster Centers:") 
print(kmeans_model$centers)

ex-9:

install.packages("ggplot2")
install.packages("caret")
library(class)      
library(ggplot2)    
library(caret)      
library(dplyr)      


data(iris)


iris$Disease <- iris$Species
iris$Species <- NULL


normalize <- function(x) {
  return ((x - min(x)) / (max(x) - min(x)))
}

iris_normalized <- as.data.frame(lapply(iris[1:4], normalize))
iris_normalized$Disease <- iris$Disease

set.seed(123)  
train_index <- createDataPartition(iris_normalized$Disease, p = 0.7, list = FALSE)
train_data <- iris_normalized[train_index, ]
test_data <- iris_normalized[-train_index, ]

k <- 5 
predicted <- knn(train = train_data[ ,1:4],
                 test = test_data[ ,1:4],
                 cl = train_data$Disease,
                 k = k)

conf_matrix <- confusionMatrix(predicted, as.factor(test_data$Disease))
print(conf_matrix)

test_data$Predicted <- predicted

ggplot(test_data, aes(x = Petal.Length, y = Petal.Width, color = Predicted, shape = Disease)) +
  geom_point(size = 3) +
  labs(title = paste("KNN Classification (k =", k, ")"),
       x = "Petal Length",
       y = "Petal Width") +
  theme_minimal()



ex-10:

install.packages("arules")
install.packages("arulesViz")
library(arules)
library(arulesViz)

data("Groceries")
summary(Groceries)


rules <- apriori(Groceries, 
                 parameter = list(supp = 0.01, conf = 0.5, minlen = 2))

summary(rules)
inspect(sort(rules, by = "lift")[1:10])  

plot(rules, method = "scatterplot", measure = c("support", "confidence"), shading = "lift")
plot(rules, method = "graph", engine = "htmlwidget")
```
